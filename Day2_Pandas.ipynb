### Reading CSV file

```python
import pandas as pd
data=pd.read_csv("train.csv")
```

```python
data
```

```python
print("\n First 5 Rows of Dataset:")
print(data.head())
```

```python
print("\nShape of Dataset:", data.shape)
```

#### Data Selection using loc and iloc


```python
# Using loc – select specific columns by names
print(" Selecting columns (Name, Age, Sex, Survived):\n")
print(data.loc[:, ['Name', 'Age', 'Sex', 'Survived']].head())
```

```python
# Using iloc – select rows and columns by index
print("\n Selecting first 5 rows and first 4 columns using iloc:\n")
print(data.iloc[0:5, 0:4])
```

```python
# Selecting passengers older than 60
print("\n Passengers aged above 60:\n")
print(data.loc[data['Age'] > 60, ['Name', 'Age', 'Sex', 'Survived']])
```

#### Data Cleaning (Handling Missing Values & Duplicates)

```python
# Checking missing values
print(" Missing Values Count:\n", data.isnull().sum())
```

```python
# Filling missing Age with mean
data['Age'].fillna(data['Age'].mean())

```

```python
# Filling Embarked with mode (most common value)
data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
# Dropping the 'Cabin' column (too many missing values)
data.drop('Cabin', axis=1, inplace=True)
```

```python
# Checking duplicates
duplicates = data.duplicated().sum()
print("\n Number of duplicate rows:", duplicates)
```

```python
# Removing duplicates if any
data.drop_duplicates(inplace=True)
```

```python
# Verify cleaning
print("\n After Cleaning:")
print(data.isnull().sum())

```

#### GroupBy and Aggregations

```python
# Group passengers by Sex and calculate survival rate
gender_group = data.groupby('Sex')['Survived'].mean()
print(" Survival Rate by Gender:\n", gender_group)
```

```python
# Group by Passenger Class and get average Age & Fare
pclass_group = data.groupby('Pclass').agg({
    'Age': 'mean',
    'Fare': ['mean', 'min', 'max'],
    'Survived': 'mean'
})
print("\n Average Age, Fare, and Survival Rate by Passenger Class:\n", pclass_group)

```

```python
# Group by Embarked (Port)
embarked_group = data.groupby('Embarked')['Survived'].agg(['mean', 'count'])
print("\n Survival Rate and Passenger Count by Port of Embarkation:\n", embarked_group)
```

#### Summary Report

```python
print(" Dataset Summary Report:")
print("Total Passengers:", data.shape[0])
print("Total Features:", data.shape[1])
print("Average Age:", round(data['Age'].mean(), 2))


# Summary statistics
print("\n Statistical Summary:\n", data.describe())

print("\n Data Cleaning and Summary Report Completed Successfully!")
```
