### Visualisation

 - Using pandas, we can get insights of data by plotting the graph
 - We use matplotlib (matlab style plotting) and seaborn (seaborn is built on matplotlib)
 - For categorical data, we use bar plot, count plot, pie plot. (Univariate analysis)
 - For numerical data, we iuse histogram or distribution plot (Univariate analysis)
 - In bivariate analysis, if x and y are continuous we use scatter plot
 - In bivariate analysis, if x or y is discrete and other is continuous, we use bar plot

##### Univariate data analysis:
- Continuous: hist plot, dist plot
- Categorical: Count plot, pie plot

##### Bivariate data analysis:
- x and y continous: Scatterplot,line plot
- x continuous and y categorical: Bar plot
- x and y categorical: Cat plot

##### Multivariate Analysis
- Pairplot

### Matplotlib

##### Matplotlib is a Python library used for data visualization. It allows you to create a wide range of static, animated, and interactive plots such as line charts, bar graphs, histograms, scatter plots, and pie charts.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.Series([1,2,3,4,5,6,7,7,8])
y = pd.Series([3,4,5,1,2,3,9,10,7])

plt.figure(figsize=(15,3))
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('Graph')

plt.plot(x,y,linestyle='dotted',marker='*', color='green')
plt.show()
```

### Seaborn

##### Seaborn is a statistical data visualization library built on top of Matplotlib.
It simplifies complex visualizations and makes them look more aesthetic, informative, and appealing with just a few lines of code.

##### Seaborn is ideal for analyzing data relationships, distributions, and correlations — especially when working with Pandas DataFrames.
