#### Random Forest (Bagging Method)

Random Forest = Many Decision Trees + Voting/Averaging

Simple Explanation

* A single decision tree is weak, overfits, and is unstable.
* So Random Forest creates many trees and combines them.

How it works (Step-by-step)

* 1️ Take the original dataset
* 2️ Create multiple bootstrap samples (random samples with replacement)
* 3️ Build one decision tree on each sample
* 4️ At each split, choose random subset of features
* 5️ Final output =
* Classification → Majority vote
* Regression → Average of all trees

#### Why it is better than a single tree?

* Reduces overfitting (major advantage)
* Reduces variance
* More accurate and stable
* Works well even with missing values
* Handles thousands of features
* Less sensitive to noise

```python
# RandomForset implementation
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    random_state=42
)

rf.fit(x_train, y_train)
r_pred=cv.predict(x_test)

print("Random Forest Accuracy:", accuracy_score(y_test, r_pred))
```

#### Gradient Boosting (Boosting Method)

Gradient Boosting = Build trees sequentially, each tree fixes mistakes of the previous one.

Simple Explanation
* Unlike Random Forest (parallel), Gradient Boosting builds trees one after another.
* Each new tree focuses on the errors (residuals) made by previous trees.

How it works

* 1️ Start with a simple model (weak learner)
* 2️ Predict output — errors happen
* 3️ Build the next tree to correct those errors
* 4️ Repeat…
* 5️ Final model = sum of all weak trees

Example (simple idea)

* Tree 1 → predicts 70% correctly
* Tree 2 → tries to fix mistakes of Tree 1
* Tree 3 → fixes mistakes of Tree 2
* Final accuracy becomes very high

Why it's powerful?

* Low bias, low error
* Works very well on structured/tabular data
* Best for small/medium datasets

Disadvantages:

* Slow training
* Can overfit if not tuned properly
* Sensitive to outliers

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

gb.fit(x_train, y_train)
gb_pred = gb.predict(x_test)

print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_pred))
```

#### XGBoost (Extreme Gradient Boosting)

Why XGBoost is better?

XGBoost adds four big improvements:

* 1. Regularization (L1 & L2)
* Controls overfitting
* (This is why XGBoost performs better than plain GradientBoosting)

* 2. Parallel processing
* Trains much faster

* 3. Tree pruning
* Uses “max_depth” + “learning_rate” + “gamma” to avoid huge trees

* 4. Handles missing values automatically
     
###### How XGBoost works (intuition)

* Same boosting idea (tree-by-tree)
BUT optimized with:
* Regularization
* Shrinkage
* Weighted quantile histogram
* Handling sparse data
* Efficient memory usage
* So it becomes fast + powerful.

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=150,
    learning_rate=0.1,
    max_depth=4,
    eval_metric='logloss'
)

xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_test)

print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
```
