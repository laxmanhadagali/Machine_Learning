```python
# import basic libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# seaborn has a inbuilt datasets
sns.get_dataset_names()
```

```python
# load iris dataset
data=sns.load_dataset("iris")
```

```python
data
```

# Relational plots
Relational plots are used to understand relationship between two variables

- line plot
- scatter plot
- relplot

```python
# line plot
sns.lineplot(data=data,x='sepal_length',y='sepal_width',color='green')
```

```python
sns.lineplot(data=data,x='sepal_length',y='sepal_width',hue='species',ci=None) #ci : confidence interval (it removes shaded region if we give it as none)
```

**scatter plot**

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length')
plt.show()
```

```python
sns.scatterplot(data=data,x='sepal_length',y='petal_length',hue='species',palette='rainbow')
plt.show()
# palette : is used to give color when we have multiple values i.e when use hue we can't use color so we use palette
```

**rel plot**

```python
# relplot by default it is scatter plot
# relplot can be converted into line plot by giving kind="line"
sns.relplot(data=data,x='petal_length',y='sepal_width')
```

```python
# relplot can be converted into line plot by giving kind="line"
sns.relplot(data=data,x='petal_length',y='sepal_width',kind='line',ci=None)
```

```python
# scatterplot just gives us the Points but
# relpot will join the points and gives as line plot
```

# Categorical plots
It is used to analyse one numerical and one categorical column

- count plot
- box plot
- barplot
- voilon plot
- strip plot
- swarn plot
- cat plot
  

**count plot**
- It just gives count of categories

```python
sns.countplot(data=data,x='species',palette='rainbow')
```

**Bar plot**

```python
#bar plot
sns.barplot(data=data,x='species',y='petal_width')
plt.show()
```

**Boxplot**

```python
# box plot is used to identify outlirrs
sns.boxplot(data=data,x='sepal_length')
plt.show()
```

```python
sns.boxplot(data=data,x='species',y='sepal_length')
# x should be a categorical column 
# y should be a numerical column
```

**violin plot**

```python
sns.violinplot(data=data,x='sepal_width')
```

```python
sns.violinplot(data=data,x='species',y='sepal_width',palette='rainbow')
```

**Strip plot**

```python
sns.stripplot(data=data,x='species',y='sepal_width',hue='species')
plt.show()
```

```python
sns.swarmplot(data=data,x='species',y='sepal_width',hue='species')
#strip overlaps
#swarmp doesnot overlap
```

**cat plot**
- using cat plot we can plot box,violin,strip,swarm

```python
sns.catplot(data=data,x='species',y='sepal_length',hue='species',kind='swarm')
plt.show()
#by default it will give scatter plot
```

# Disribution plots
- histplot
- distplot
- #these two graps used to check distribution of data

```python
sns.histplot(data=data,x='petal_width',kde=True,hue='species')
#to get a curve use kde=True
```

```python
#didtplot
sns.distplot(x=data['sepal_length'])
plt.show()
```

# PairPlot

```python
#used to understand relationship between multiple plots
sns.pairplot(data=data,hue='species')
plt.show()
```
